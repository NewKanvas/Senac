## Atividade - Walmart (1 momento)

Apresente um subplots de grficos de temperaturas semanais das lojas 40, 41, 42 e 43 do Walmart filtrando pelo ano de 2010

Apresente um subplots de grficos de linhas das vendas feitas semanalmente (Weekly_Sales) pelas lojas 40, 41, 42 e 43 do Walmart filtrando pelo ano de 2010. Para facilitar a visualizao, normalize os dados de vendas e ajuste o grfico para que os subplots compartilhem os mesmos eixos (x e y)

```python
df = pd.read_csv('../csv/Walmart_Store_sales.csv')
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
df.head(3)
```

```python
df2 = df[(df['Date'].dt.year == 2010) & (df['Store'].isin([40, 41, 42, 43]))]
df2
```

```python
df2.loc[df2['Store'] == 40]['Temperature'].head()
```

```python
df2.loc[df2['Store'] == 40]['Temperature'].plot()
```

```python
plt.subplot(1, 4, 1)
plt.plot(df2.loc[df2['Store'] == 40]['Temperature'])

plt.subplot(1, 4, 2)
plt.plot(df2.loc[df2['Store'] == 41]['Temperature'])

plt.subplot(1, 4, 3)
plt.plot(df2.loc[df2['Store'] == 42]['Temperature'])

plt.subplot(1, 4, 4)
plt.plot(df2.loc[df2['Store'] == 43]['Temperature'])
```

```python
filtered_data = df.loc[(df['Date'].dt.year == 2010) & (df['Store'].isin([40, 41, 42, 43]))]

# Criando um subplot
fig, axs = plt.subplots(4, 1, figsize=(10, 10), sharex=True)

# Iterando sobre as lojas e plotando o grfico de temperatura semanal para cada uma
for i, store_num in enumerate([40, 41, 42, 43]):
    ax = axs[i]
    store_data = filtered_data.loc[filtered_data['Store'] == store_num]
    ax.plot(store_data['Date'], store_data['Temperature'], label=f'Store {store_num}')
    ax.set_ylabel('Temperature')
    ax.legend()
```
