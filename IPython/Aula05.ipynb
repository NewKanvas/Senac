### **Carregando os dados**

```python
url = 'https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv'
titanic = pd.read_csv(url)
titanic
```

```python
# 1. Criar uma cópia do DataFrame original com apenas as 5 primeiras linhas;

titanic2 = titanic.head(5)
titanic2
```

```python
# 2. Remover a coluna Fare, que possui o valor das tarifas cobradas para cada passageiro;

del titanic2['Fare']
titanic2
```

```python
# 3. Remover a linha, pelo índice, contendo as informações da passageira “Miss. Laina Heikkinen”
i = titanic2.loc[(titanic2['Name'] == "Miss. Laina Heikkinen")]
titanic2.drop(i.index, inplace=True)
titanic2
```
