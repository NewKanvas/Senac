## Pandas DataFrame - Variância
● A variância é uma medida que expressa quanto os dados de um conjunto
estão afastados de seu valor esperado;
● Simplificando, seria basicamente a distância que um determinado valor do
conjunto em relação a média;
● Dentro da biblioteca pandas, a variância pode ser calculada tanto para series
como para dataframes através da função var();

## Pandas DataFrame - Desvio Padrão
● Se observarmos a função de variância vemos que o resultado (σ²) é obtido
com a variável elevada ao quadrado. Se tirarmos a raiz quadrada desse valor
obteremos o desvio padrão;
● O desvio padrão é uma medida do nível de dispersão, isto é, ele indica quão
uniforme está um conjunto de dados.
● Ou seja, quanto maior o desvio padrão, o conjunto de dados está mais
distante da média. Quanto mais próximo de 0 ele estiver, temos o conjunto
de dados mais próximo à média;

```python
#Variancia 
print(series_um.var())

print(series_dois.var())

print(df_treino['Age'].var())
```

# Calculando o desvio padrao

```python
print(np.sqrt(series_um.var()))
```

```python
print(np.sqrt(series_um.var(ddof = 0)))
```

```python
print(np.std(series_um, ddof = 1))
```

```python
# Exemplo com DataFrame
df = pd.DataFrame({'coluna': [1, 2, 3, 4, 5]})

# Calculando variância e desvio padrão
variance = df['coluna'].var()
std_deviation = df['coluna'].std()

print(f"Variância: {variance}")
print(f"Desvio Padrão: {std_deviation}")
```

## Pandas DataFrame - Coeficiente de Variação
● O coeficiente de variação permite analisarmos conjuntos de dados em relação ao
valor médio, mesmo que tenhamos unidades de medidas diferentes;
● O coeficiente de variação é calculado a partir do desvio padrão pela fórmula: CV
= (s/x̅) x 100
● Onde,
● s → é o desvio padrão
● x̅ → é a média dos dados
● CV → é o coeficiente de variação
● Como o resultado do coeficiente de variação é dado em porcentagem, facilita a
interpretação do resultado e comparação com outros conjuntos de dados.

Seja um dado X e um dado Y, onde podem ser enxergados como lista. O coeficiente de variação analisa a relação entre o conjunto de dados mencionados. Ao multiplicar por 100, você tem esse resultado em porcentagem

```python
#Coeficiente de variação
#Calculo = (Desvio Padrão / Média ) * 100
print((np.sqrt(series_um.var()) / series_um.mean()) * 100)
print((np.sqrt(series_dois.var()) / series_dois.mean()) * 100)
```


## Pandas DataFrame - Correlação
● Devido a covariância às vezes se tornar um pouco difícil de ser compreendida e
comparada e com a necessidade de normalizar seu resultado para torná-lo mais
interpretável é que se utiliza a correlação;
● A correlação faz a mesma análise da covariância, porém delimitando os valores sempre
entre -1 e 1;
● O valor -1 indica a anticorrelação perfeita e 1 indica a correlação perfeita;
● Dentro da biblioteca pandas, a variância pode ser calculada tanto para series como para
dataframes através da função corr().

```python
numeric_columns = df_treino.select_dtypes(include=[np.number]).columns
df_treino[numeric_columns].corr()
```

## Pandas DataFrame - Covariância
● A covariância mede a relação linear entre duas variáveis;
● Ela tenta medir o afastamento dos valores de duas variáveis em relação às suas
médias;
● Uma covariância igual a 0 indica que as duas variáveis são totalmente
independentes, enquanto que uma covariância alta e positiva indica que uma
variável é grande quando a outra é grande;
● Uma covariância negativa e com valor absoluto alto indica que uma variável é
pequena quando a outra é grande;
● Dentro da biblioteca pandas, a variância pode ser calculada tanto para series como para
dataframes através da função cov().
