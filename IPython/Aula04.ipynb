### **Consultas por Fatiamento**

#### Fatiamento para exibir as primeiras 10 linhas da coluna 'Weekly_Sales'

```python
walmart[:10]['Weekly_Sales']
```

A expressão walmart.loc[0:10]['Date'] está sendo usada para selecionar as datas na coluna 'Date' do DataFrame 'walmart' para as linhas de 0 a 10, inclusive. Em resumo, ela extrai um subconjunto específico de datas do início ao décimo elemento do DataFrame.

```python
walmart[:1]['Date']
```

```python
walmart[1:5]
```

### **Consultas por Indice**

```python
walmart.loc[0]['Date']
```

#### Usando loc para obter o valor na posição (linha 0, coluna 1)
Loc pega o indice da linha, o ILOC pega linha e a coluna [linha ,coluna]

```python
walmart.iloc[0,1]
```

```python
walmart.loc[0]
```

```python
walmart.loc[[1,2]]
```

### Consulta com criterios

```python
walmart.loc[walmart['Holiday_Flag'] == 1]
```

```python
walmart.loc[walmart['Temperature']>90]
```

### Consulta com mais de um criterio

```python
walmart.loc[walmart['Temperature']>90 & (walmart['Holiday_Flag'] == 1)]
```

### **Consulta por Query**

```python
walmart.query('Holiday_Flag == 1')
```

#### Consulta por Query com mais de um criterio

```python
walmart.query('20 <= Temperature < 50')
```

```python
walmart.query('Store  in [1,3]')
```

```python
tempe = 50.0
walmart.query('Temperature > @tempe')
```
