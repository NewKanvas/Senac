```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1,2,3]})

right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

merge tem uma opção suffixes para especificar strings a serem concatenadas nos nomes que se sobrepõem, nos objetos DataFrame à esquerda e à direita

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

# Fazendo mergem com base no indice

Voce pode passar lef_index = True ou right_index=True ( ou ambos) para informar que o indice devera ser usado como a chave do merge

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})

right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
left1.join(right1, on='key', how='outer')
```

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
```

- Por padrao,concat atua em aaxis=0, gerenado outra Series.
- Se voce passar axis=1, o resultado sera um dataframe.

```python
pd.concat([s1,s2,s3])
```

- Não há sobreposição no outro eixo, o que gera a união ordenada (a junção 'outer') dos índices
- Podemos fazer a sua intersecção passando join = 'inner'

```python
s4 = pd.concat([s1, s3])
s4
```

```python
s4 = pd.concat([s1, s3])
s4
```

- Os rotulos 'f' e 'g' desapareceram por causa da opçao join='inner'
- Suponha que quisessemos criar m indeice hieraquivo no eixo  da concaneta [...]

```python
pd.concat([s1, s4], axis=1, join='inner')
```

```python
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three'])
result
```

```python
result.unstack()
```

No caso da combinação de Series ao longo de axis=1, as keys passam a ser os cabeçalhos das colunas do DataFrame

```python
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three'])
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'], columns=['one', 'two'])
df1
```

```python
df2 = pd.DataFrame(5 + np.arange(4).reshape(2,2), index=['a', 'c'], columns=['three', 'four'])
df2
```

```python
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'])
```

- Quando o indice das linhas dos DataFrames não contem nenhum dado relevante, podemos passar ignore_index = True

## Combinando dados om sobreposição

Podemos ter dois conjuntos de dados cujos indices se sobreponham de forma total ou parcial.

```python
a = pd.Series([np.nan, 2.5, 0, 3.5, 4.5, np.nan], index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series([0, np.nan, 2, np.nan, np.nan, 5], index=['a', 'b', 'c', 'd', 'e', 'f'])
```

```python
a
```

```python
b
```

```python
b.combine_first(a)
```

- Com DataFrames, combine_first faz o mesmo coluna a coluna

```python
df1 = pd.DataFrame({'a': [1, np.nan, 5, np.nan],
                    'b': [np.nan, 2, np.nan, 6],
                    'c': range(2,18,4)})
df1
```

```python
df2 = pd.DataFrame({'a': [5, 4, np.nan, 3, 7],
                    'b': [np.nan, 3, 4, 6, 8]})
df2
```

```python
df1.combine_first(df2)
```
