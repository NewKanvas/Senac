```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        ax[i,j].plot(np.random.randn(15).cumsum())
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()  # Cria uma figura e os eixos
ax.plot(x, x, label='linear')  # Plotando dados
ax.plot(x, x**2, label='quadratico')  # Plotando mais dados
ax.plot(x, x**3, label='cubico')  # ... e mais.
ax.set_xlabel('eixo x')  # Adicionando rtulo no eixo x.
ax.set_ylabel('eixo y')  # Adicionando rtulo no eixo y.
ax.set_title("Plotando linhas")  # Adicionando um ttulo.
ax.legend()  # Adicionando uma legenda.
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Vendas")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Faturamento")

plt.suptitle("Loja XPTO")
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Gasto de calorias", fontdict = font1)
plt.xlabel("Tempo em segundos", fontdict = font2)
plt.ylabel("Calorias", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Gasto de calorias", loc = 'left')
plt.xlabel("Tempo em segundos")
plt.ylabel("Calorias")

plt.plot(x, y)
plt.show()
```

Para alterar os tiques do eixo x, a opo mais simples  usar set_xticks e set_xticklabels. A primeira instrui a matplotlib acerca do local para posicionar os tiques no intervalo de dados; por padro, esses locais tambm sero os rtulos.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
ax.set_title("Plotagem de frequencia")
ax.set_xlabel('Estgios')
```
